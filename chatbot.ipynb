```python
import json 
import numpy as np 
import pandas as pd
import tensorflow as tf
import json 
import random
import pickle
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

Address receive complaints<br>
Prepare and load your data:    

```python
intents = pd.read_json("priti_intents.json")
print( intents)
```

Write code to read your json file, and store the elements into a group of lists

```python
patterns = []
tags = []
labels = []
responses = []
```

```python
for intent in intents['intents']:
    for pattern in intent['patterns']:
        patterns.append(pattern)
        tags.append(intent['tag'])
    responses.append(intent['responses'])
    
    if intent['tag'] not in labels:
        labels.append(intent['tag'])
        
num_classes = len(labels)
```

Encode the list of intents, these will be the classes for your model in total 8 classes.

```python
label_encoder = LabelEncoder()
label_encoder.fit(tags)
tags = label_encoder.transform(tags)
```

Use keras tokenizer to tokenize the patterns, then carryout   the necessary pre-
processing steps that you think are required to prepare the data for a deep learning 
training on the intent patterns in order to identify intents.

```python
vocab_size = 1000
embedding_dim = 16
max_len = 20

tokenizer = Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(patterns)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(patterns)
padded_sequences = pad_sequences(sequences, truncating='post', maxlen=max_len)
```

```python
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim, input_length=max_len))
model.add(GlobalAveragePooling1D())
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
epochs = 300
history1 = model.fit(padded_sequences, np.array(tags), epochs=epochs)
```

```python
epochs = 500
history2 = model.fit(padded_sequences, np.array(tags), epochs=epochs)
```

```python
# to save the trained model
model.save("NLP_Chatmodel")

# to save the fitted tokenizer
with open('tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle)
    
# to save the fitted label encoder
with open('label_encoder.pickle', 'wb') as ecn_file:
    pickle.dump(label_encoder, ecn_file)
```

```python
 # load trained model
with open("sample_intents.json") as file:
    data = json.load(file)
def saved_pickels():
    
        
    # load trained model    
    model = keras.models.load_model('NLP_Chatmodel')

    # load tokenizer object
    with open('tokenizer.pickle', 'rb') as handle:
        tokenizer = pickle.load(handle)

    # load label encoder object
    with open('label_encoder.pickle', 'rb') as enc:
        label_encoder = pickle.load(enc)
        
```

```python

def chat_bot():
    
    #loading saved files
    saved_pickels()

    # parameters
    max_len = 20
    
    while True:
        
        print("User: " ,end="")
        inp = input()
        if inp.lower() == "quit":
            break

        result = model.predict(keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences([inp]),
                                             truncating='post', maxlen=max_len))
        tag = label_encoder.inverse_transform([np.argmax(result)])

        for i in data['intents']:
            if i['tag'] == tag:
                print("Priti:" ,np.random.choice(i['responses']))
                
               
print("Type quit to stop chatting!")
print("Priti: " ,"Hi! How can I help?") 
chat_bot()
```
