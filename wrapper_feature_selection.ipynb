```python
import warnings
import gc
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataframe = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Predictive_Data/alldata (1).csv")
```

```python
dataframe.info()
```

```python
df = dataframe.sample(frac=0.25)
df.info()
```

```python
df.head(15)
```

```python
df.columns
```

```python
df = df._convert(numeric=True)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
cormat = df.corr()
round(cormat,3)
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Total number of missing values
df.isnull().sum().sum()
```

```python
X = df.drop('marker', axis=1)
X = X.fillna(X.mean())
print("total null values", X.isnull().sum().sum())
X.shape
```

```python
y = df['marker']
y = y.fillna(y.mode().iloc[0])
print("total null values", y.isnull().sum().sum())
y.shape
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state = 40)
X_train.shape, y_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, max_depth=10)
```

```python
from sklearn.feature_selection import RFE
# here we want only 50 final feature, we do this to produce a ranking
n_features_to_select = 50
rfe = RFE(regressor, n_features_to_select=n_features_to_select)
rfe.fit(X_train, y_train)
```

```python
from operator import itemgetter
features = X_train.columns.to_list()
for x, y in (sorted(zip(rfe.ranking_ , features), key=itemgetter(0))):
    print(x, y)
```

```python
predictions = rfe.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import math
MSE = MSE = mean_squared_error(y_test,predictions)
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

```python
target     = 'marker'
```

```python
output = pd.DataFrame({"Id":X_test.index, target:predictions})
output.to_csv('submission.csv', index=False) 
```
